### Gather:
df_twitter_archive are downloaded manually from the website. And then I used the tweet_id pulled from the csv file I gathered to use tweepy library to query twitter API which let me to get the json data regarding each tweet.

And the information I want is favourite count and retweet count. I first write the json data using json library to a txt file and separated by \n so every line represent a tweet. And then I use json library to decode the json data and extracted each tweet's id, fav count and retweet count and make it a dictionary to be appended to a list so I can load it to a pandas dataframe. (later loaded as df_tweet_count)

The last file is downloaded programmatically using request library. After getting the response, the content is written to a tsv file. (later loaded as df_image_pred)
